## Create comparison images by aligning pairwise to the uniprot structures

```python
x = [1, 0, 0]
y = [0, 1, 0]
z = [0, 0, 1]
```

```python
for count, pdb_id in enumerate(pdbs):
    cmd.delete("all")
    print(pdb_id)
    print("progress: "+str(count)+" out of "+str(len(pdbs)))
    uniprot_id = translator["uniprot"][pdb_id]
    pdam_id = translator["pdam"][pdb_id]

    path1 = os.path.join("model_structures", pdb_id+".pdb")
    path2 = os.path.join("edited_alphafold_structures", uniprot_id+".pdb")
    path3 = os.path.join("edited_pdam_structures", pdam_id+".pdb")
    
    cmd.load(path1)
    cmd.load(path2)

    cmd.align(pdb_id, uniprot_id)
    cmd.color("red", pdb_id)
    cmd.color("green", uniprot_id)

    cmd.bg_color("white")

    cmd.png(os.path.join("comparison_images", pdb_id+"model-vs-uniprot"), 1000, 800, dpi=150, ray=1)

    cmd.rotate(x, 180)
    cmd.png(os.path.join("comparison_images", pdb_id+"model-vs-uniprot-Xrotated"), 1000, 800, dpi=150, ray=1)
    
    cmd.delete("all")
    
    cmd.load(path2)
    cmd.load(path3)
    
    cmd.align(pdam_id, uniprot_id)
    cmd.color("green", uniprot_id)
    cmd.color("blue", pdam_id)
    
    cmd.bg_color("white")

    cmd.png(os.path.join("comparison_images", pdb_id+"uniprot-vs-pdam"), 1000, 800, dpi=150, ray=1)

    cmd.rotate(x, 180)
    cmd.png(os.path.join("comparison_images", pdb_id+"uniprot-vs-pdam-Xrotated"), 1000, 800, dpi=150, ray=1)
    
```
