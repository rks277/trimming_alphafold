Now we can use this to delete any residues on strings, an example run is below.

```python
structure = PDBParser().get_structure("P49410", "P49410.pdb")
```

```python
CUTOFF = 5
DISTANCE = 10

structure_copy = structure
model = structure_copy[0]

for chain in model:
    for residue in list(chain):
        #print(residue.get_id()[1])
        string = True
        center_atoms = Selection.unfold_entities([chain[residue.get_id()[1]]], 'A')
        atom_list = [atom for atom in structure.get_atoms()]
        ns = NeighborSearch(atom_list)
        nearby_residues = {res for center_atom in center_atoms
                           for res in ns.search(center_atom.coord, DISTANCE, 'R')}
        nearby_residues = [res.id[1] for res in nearby_residues]
        #print(str(residue.get_id()[1])+" is near " + str(nearby_residues))
        for i in nearby_residues:
            if abs(i - residue.get_id()[1]) > CUTOFF:
                string = False
        if string:
            #print(str(residue.get_id()[1])+" is on a string!")
            chain.detach_child((' ', residue.get_id()[1], ' '))
```

```python
for count, pdb_id in enumerate(pdbs):
    print(str(pdb_id)+", progress: "+str(count)+" out of "+str(len(pdbs)))
    uniprot_id = translator["uniprot"][pdb_id]
    pdam_id = translator["pdam"][pdb_id]

    path1 = os.path.join("model_structures", pdb_id+".pdb")
    path2 = os.path.join("alphafold_structures", uniprot_id+".pdb")
    path3 = os.path.join("pdam_structures", pdam_id+".pdb")
    
    alphafold_structure = PDBParser().get_structure(uniprot_id, path2)
    pdam_structure = PDBParser().get_structure(pdam_id, path3)
    
    alphafold_copy = alphafold_structure
    alphafold_model = alphafold_copy[0]

    for chain in alphafold_model:
        for residue in list(chain):
            
            string = True
            center_atoms = Selection.unfold_entities([chain[residue.get_id()[1]]], 'A')
            atom_list = [atom for atom in structure.get_atoms()]
            ns = NeighborSearch(atom_list)
            nearby_residues = {res for center_atom in center_atoms
                               for res in ns.search(center_atom.coord, DISTANCE, 'R')}
            nearby_residues = [res.id[1] for res in nearby_residues]
            
            for i in nearby_residues:
                if abs(i - residue.get_id()[1]) > CUTOFF:
                    string = False
            if string:
                chain.detach_child((' ', residue.get_id()[1], ' '))
                
    io.set_structure(alphafold_model)
    io.save("edited_alphafold_structures/"+uniprot_id+".pdb")
    
    pdam_copy = pdam_structure
    pdam_model = pdam_copy[0]

    for chain in pdam_model:
        for residue in list(chain):
            
            string = True
            center_atoms = Selection.unfold_entities([chain[residue.get_id()[1]]], 'A')
            atom_list = [atom for atom in structure.get_atoms()]
            ns = NeighborSearch(atom_list)
            nearby_residues = {res for center_atom in center_atoms
                               for res in ns.search(center_atom.coord, DISTANCE, 'R')}
            nearby_residues = [res.id[1] for res in nearby_residues]
            
            for i in nearby_residues:
                if abs(i - residue.get_id()[1]) > CUTOFF:
                    string = False
            if string:
                chain.detach_child((' ', residue.get_id()[1], ' '))
                
    io.set_structure(pdam_model)
    io.save("edited_pdam_structures/"+pdam_id+".pdb")
```
